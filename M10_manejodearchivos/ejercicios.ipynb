```python
import random;
import os;

numeros = [];
pair = [];
contador = 0;

while contador < 200:
    number = random.randint (1, 500);
    contador += 1;
    pair = [contador, number]
    numeros.append(pair);


archivo = open(ubicacion+"pairs_numbers.txt", "w");
for index, elemento in enumerate (numeros):
    archivo.write(str(index+1)+"    "+str(elemento[1])+"\n");

archivo.close();
print("Se ha creado un nuevo archivo.");

print(numeros);
print(len(numeros));
```

```python
import os;
archivo = open(ubicacion+"pairs_numbers.txt", "r");
array = [];

for linea in archivo:
    linea = linea.rstrip();
    linea_limpia = linea.replace("    ","-");
    array.append(linea_limpia.split("-"));

index = 0;
array_interno = [];
new_array = [];

for elemento in array:
    while index < 1:
        for string in elemento:
            numero = int(string);
            if (len(array_interno) < 2):
                array_interno.append(numero);
            else:
                new_array.append(array_interno);
                array_interno = [];
                array_interno.append(numero);
        index += 1;
        # print(new_array);
    index = 0;

archivo.close();

print (new_array);
print (len(new_array));
```
